### 2. Variable scope

In Python, the scope of a variable refers to the region of the program where the variable is defined and can be accessed. There are two types of scope in Python: local and global.

A **local variable** is defined inside a function and has a local scope. It can only be accessed within that function and is not available outside of it.

A **global variable** is defined outside of any function and has a global scope. It can be accessed from any part of the program. However, if a local variable with the same name is defined within a function, it will take precedence over the global variable and its value will be used instead.

It's important to understand the scope of variables in Python to avoid naming conflicts and to write efficient and maintainable code.

### 4. Methods and formulas

Below are two weights (in kilograms) of some apples given, along with the prices in euros. We will calculate the price per kilogram by taking the difference in price and divide it by the difference in weight.

```python
# define a list of apple weights and apple prices
weights = [1.50, 2]
prices = [2.97, 3.96]


def price_per_weight(weights, prices):
    a = (prices[1] - prices[0])/(weights[1] - weights[0])
    return round(a,2)


price_per_weight(weights, prices)
```

We can use the price per weight now to determine how much 34 kilograms of apples would cost.

```python
price = price_per_weight(weights, prices)

# calculate the costs of 34 kilograms
price * 34
```

Basically, we calculated a linear formula. We could also have a method for linear formulas and apply the same input.

```python
def linear_formula_slope(x, y):
    a = (y[1] - y[0])/(x[1] - x[0])
    return a


linear_formula_slope(weights, prices)
```

### 5. Working with collections (lists/arrays)

To create a list, you use the square brackets and add elements into it accordingly. If you do not pass any elements inside the square brackets, you get an empty list as the output.

    my_list = []

    my_list = [1, 2, 3, 'example', 3.132] #creating list with data

You have several other functions that can be used when working with lists.

- len() function returns to us the length of the list.
- index() function finds the index value of value passed where it has been encountered the first time.
- count() function finds the count of the value passed to it.

See more: https://www.edureka.co/blog/data-structures-in-python/

```python
def fit(X, y):

    # compute the mean
    X_mean = mean(X)
    y_mean = mean(y)

    num = 0
    den = 0
    
    # least squared method to get the weight and bias
    for i in range(len(X)):
        num = num + (X[i] - X_mean) * (y[i] - y_mean)
        den = den + (X[i] - X_mean) ** 2
        
    weights = num / den
    bias = y_mean - weights * X_mean

    return weights, bias
```

# AI 

```python
#arrange data in (X,y) tuple 

X=data.Height.values.reshape(-1,1)
y=data.Weight.values.reshape(-1,1)

X,y
```

Furthermore, it is indicated you scale/normalize the data! 

### 3. Data preparation

Train/test split, to be able to validate the performance of a model.

The training set is used for training the model, and the testing set is used to test your model.

Think of a student attending a math course, practicing on a set of math exercices; afterwards showing expertise in a test on similar math exercies that the practical was done.


```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
x_train.shape,x_test.shape,y_train.shape, y_test.shape
```

### 4. Training model

Machine Learning community has already programmed the most popular algorithms. You will be able to use these algorithms in your notebooks as follows:

1. Import the algorithm library (e.g. `from sklearn.linear_model import LinearRegression` or `from sklearn.linear_model import LogisticRegression`)
2. Define a variable for your algorithm (e.g. `lm = LinearRegression()` or `lgm = LogisticRegression()`)
3. Fit a model using the algorithm (e.g. `lm` or `lgm`) and the data you prepared (e.g. `x_train` and `y_train`)

We call the method `fit()`(see https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html).

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train)
```

### 5. Predicting

We call the method `predict()`(see https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html).

```python
x_test, y_test
```

```python
y_pred=lm.predict(x_test)
y_pred
```

NOTE: You always will predict using data you did not train your model with, in this case `x_test`.

### 6. Validate model

Look at performance scores, argue the performance of your model; go back to step 2. Data collection and analysis and start all over again; try new things.

First we are going to evaluate the performance of the model using the acurracy score. We call the method score. (see https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html)

NOTE: this score is used only for classification problems. 

```python
# prediction accuracy 
lm.score(x_test,y_test)*100
```

Notice we use the test dataset, i.e. `x_test` and `y_test`, to evaluate the performance of the model.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mean_absolute_error(y_test,y_pred),np.sqrt(mean_squared_error(y_test,y_pred))
```
